```python
zomato['cost'] = zomato['cost'].astype(str) #Changing the cost to string
zomato['cost'] = zomato['cost'].apply(lambda x: x.replace(',','.')) #Using lambda function to replace ',' from cost
zomato['cost'] = zomato['cost'].astype(float) # Changing the cost to Float
zomato.info()
```

```python
#Reading Rate of dataset
zomato['rate'].unique()
```

```python
#Removing '/5' from Rates
# Remove rows where rate is 'NEW' or '-'
zomato = zomato.loc[zomato.rate != 'NEW']
zomato = zomato.loc[zomato.rate != '-'].reset_index(drop=True)

# Define a function to remove '/5' from the rate column
remove_slash = lambda x: x.replace('/5', '') if isinstance(x, str) else x

# Apply the function, strip any extra whitespace, and convert to float
zomato.rate = zomato.rate.apply(remove_slash).str.strip().astype('float')

# Display the first few values of the rate column
zomato['rate'].head()
```

```python
## Lower Casing
zomato["reviews_list"] = zomato["reviews_list"].str.lower()
zomato[['reviews_list', 'cuisines']].sample(5)
```

```python
## Removal of Puctuations
import string
PUNCT_TO_REMOVE = string.punctuation
def remove_punctuation(text):
    """custom function to remove the punctuation"""
    return text.translate(str.maketrans('', '', PUNCT_TO_REMOVE))

zomato["reviews_list"] = zomato["reviews_list"].apply(lambda text: remove_punctuation(text))
zomato[['reviews_list', 'cuisines']].sample(5)
```

```python
## Removal of Stopwords
import nltk

# Download the stopwords resource
nltk.download('stopwords')

from nltk.corpus import stopwords
STOPWORDS = set(stopwords.words('english'))
def remove_stopwords(text):
    """custom function to remove the stopwords"""
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])

zomato["reviews_list"] = zomato["reviews_list"].apply(lambda text: remove_stopwords(text))
```

```python
import re
## Removal of URLS
def remove_urls(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r'', text)

zomato["reviews_list"] = zomato["reviews_list"].apply(lambda text: remove_urls(text))
zomato[['reviews_list', 'cuisines']].sample(5)
```

```python
# Create a weighted rating feature
zomato['weighted_rating'] = zomato['rate'] * zomato['votes'] / zomato['votes'].max()
zomato.head()
```

```python
# Convert 'online_order' and 'book_table' into binary features
zomato['online_order'] = zomato['online_order'].apply(lambda x: 1 if x == 'Yes' else 0)
zomato['book_table'] = zomato['book_table'].apply(lambda x: 1 if x == 'Yes' else 0)
zomato.head()
```

```python
# Aggregate ratings and votes by location (neighborhood)
location_stats = zomato.groupby('location').agg({
    'rate': 'mean',   # Average rating for each location
    'votes': 'sum'    # Total votes for each location
}).reset_index()

# Merge aggregated statistics back into the original dataframe
zomato = zomato.merge(location_stats, on='location', suffixes=('', '_location'))

# You now have location-based mean ratings and total votes
```

Summary of Added Features:
weighted_rating: A weighted score based on both the restaurant's rating and the number of votes.
online_order and book_table: Binary indicators (1 or 0) for the availability of online ordering and table booking.
rate_location and votes_location: Aggregated statistics for the neighborhood, showing the average rating and the total number of votes in each location.

```python
from flask import Flask, request, jsonify
from transformers import AutoTokenizer, AutoModelForCausalLM
from sentence_transformers import SentenceTransformer
import torch
import json

# Initialize models
model_name = "HuggingFaceH4/zephyr-7b-beta"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name).to("cuda" if torch.cuda.is_available() else "cpu")
enc_model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')

app = Flask(__name__)

def extract_entities(ip_prompt):
    total_prompt = f"""User Query: {ip_prompt}\n\n\
    Extract these entities:
    1. name
    2. type
    3. cuisines
    4. online_order
    5. book_table
    6. location

    Return a dictionary with these keys. Use None for missing keys and "current" for unspecified addresses."""

    messages = [
        {"role": "system", "content": "Extract entities and convert to a dictionary."},
        {"role": "user", "content": total_prompt},
    ]

    encoded_input = tokenizer(messages, return_tensors="pt", padding=True).to(model.device)
    generated_ids = model.generate(encoded_input['input_ids'], max_new_tokens=300, do_sample=True, temperature=0.7, top_k=50, top_p=0.95)
    decoded_output = tokenizer.decode(generated_ids[0], skip_special_tokens=True)

    # Process LLM output into dictionary
    start_idx = decoded_output.find('{')
    json_str = decoded_output[start_idx:].rstrip('</s>').replace("None", "null")
    return json.loads(json_str)

def create_text_vector(output_dict):
    text_vector_parts = []
    for key, val in output_dict.items():
        if val is None or key == "location":
            continue
        if isinstance(val, list):
            text_vector_parts.append(f"{key}: {', '.join(val)}")
        else:
            text_vector_parts.append(f"{key}: {val}")
    return "[SEP]".join(text_vector_parts), output_dict.get("location", "current")

def search_database(vector):
    # Placeholder for actual vector search call
    collection_name = "restaurants"
    results = qdrant_client.search(
        collection_name=collection_name,
        query_vector=vector,
        limit=5
    )
    return results

@app.route('/chat', methods=['POST'])
def chat():
    user_input = request.json.get('message')
    
    # Step 1: Extract structured data from user query
    output_dict = extract_entities(user_input)

    # Step 2: Process structured data into a text vector
    text_vector, address = create_text_vector(output_dict)
    
    # Step 3: Convert to vector and perform vector search
    query_vector = enc_model.encode(text_vector)
    results = search_database(query_vector)

    # Format response
    response = {
        "user_query": user_input,
        "structured_data": output_dict,
        "recommendations": [result['name'] for result in results]
    }
    return jsonify(response)

if __name__ == '__main__':
    app.run(host='0.0.0.0', port=5000)
```
